```python
import pandas as pd
```

```python
df=pd.read_csv("/content/movie_revenues.csv")
```

```python
df.head(2)
```

```python
df.revenue.describe()
```

```python
df['revenue_min']=df['revenue'].apply(lambda x:x/1000000)
```

```python
df.revenue.describe()
```

```python
df.head(2)
```

```python
_,mean,std,*_=df.revenue_min.describe()
```

```python
mean,std
```

```python
def get_z_score(value,mean,std):
  return (value-mean)/std
```

```python
df['z_score']=df.revenue_min.apply(lambda x: get_z_score(x,mean,std))
df.head(3)
```

```python
df[df.z_score>3]
```

```python
def get_mad(s):
  median=np.median(s)
  diff=abs(s-median)
  MAD=np.median(diff)
  return MAD
```

```python
MAD=get_mad(df.revenue_min)
median=np.median(df.revenue_min)
MAD,median
```

```python
def get_modified_z_score(x,median,MAD):
  return 0.6745*(x-median)/MAD
```

```python
get_modified_z_score(2711,median,MAD)
```

```python
df['mod_z_score']=df.revenue_min.apply(lambda x:get_modified_z_score(x,median,MAD))
df.head(3)
```

```python
df[df.mod_z_score>3.5]
```
